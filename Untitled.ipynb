```python
import pandas as pd
```

```python
df  = pd.read_csv('asylum_seekers.csv')
df.drop('Unnamed: 0', axis =1 , inplace= True)
```

```python
df.rename(columns={'Country / territory of asylum/residence': 'country_of_asylum','RSD procedure type / level': 'rsd_procedure_type',
               'Tota pending start-year':'total_pending_start_year', 'of which UNHCR-assisted(start-year)':'num_unhcr_assisted_start_year' ,
                'Applied during year':'applied_during_year', 'Otherwise closed': 'otherwise_closed',
                  'Total decisions': 'total_decisions','Total pending end-year' : 'total_pending_end_year',
                   'Rejected' : 'rejected', 'Origin': 'origin','Year':'year',
                  'of which UNHCR-assisted(end-year)': 'num_unhcr_assisted_end_year'}, inplace=True)
```

```python
#filling nulls
df['rsd_procedure_type'].fillna('-', inplace = True)
```

```python
for col in ['total_pending_start_year', \
            'num_unhcr_assisted_start_year', \
            'applied_during_year', \
            'decisions_recognized', \
            'decisions_other', \
            'rejected', \
            'otherwise_closed', \
            'total_decisions', \
            'total_pending_end_year', \
            'num_unhcr_assisted_end_year']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].astype('float')
df
```

```python
df.fillna(0,inplace=True)
```

```python
## making a sum dataframe of the data:
temp = df
temp.drop('rsd_procedure_type',axis =1 , inplace = True )
temp
```

```python
origins = temp['origin'].unique().tolist()
```

```python
by_origin = pd.DataFrame(columns=['country','origin_total_applied', 'origin_recognized','origin_decisions_other','origin_rejected','origin_otherwise_closed','origin_total_decisions'])
for o in origins:
    country = temp[temp['origin'] == o]
    by_origin = by_origin.append({'country':o,'origin_total_applied':country['applied_during_year'].sum(), 'origin_recognized':country['decisions_recognized'].sum(),
                                 'origin_decisions_other': country['decisions_other'].sum(),'origin_rejected':country['rejected'].sum(),'origin_otherwise_closed':country['otherwise_closed'].sum(),
                                 'origin_total_decisions':country['total_decisions'].sum()},ignore_index=True)
by_origin
```

```python
assylum_contries = pd.DataFrame(columns=['country','in_country_total_applied', 'in_country_recognized','in_country_decisions_other','in_country_rejected','in_country_otherwise_closed','in_country_total_decisions'])
for o in origins:
    country = temp[temp['country_of_asylum'] == o]
    assylum_contries = assylum_contries.append({'country':o,'in_country_total_applied':country['applied_during_year'].sum(), 'in_country_recognized':country['decisions_recognized'].sum(),
                                 'in_country_decisions_other': country['decisions_other'].sum(),'in_country_rejected':country['rejected'].sum(),'in_country_otherwise_closed':country['otherwise_closed'].sum(),
                                 'in_country_total_decisions':country['total_decisions'].sum()},ignore_index=True)
display(assylum_contries)
```

```python
merged = pd.merge(by_origin, assylum_contries, on="country")
merged
```

```python
merged.to_csv('refugees_applications_recognition.csv',index = False)
```

```python
import pandas as pd
time_series = pd.read_csv('time_series.csv')
time_series = time_series[time_series['Year']>=2000]
```

```python
countries_asylum = time_series['Country / territory of asylum/residence'].unique().tolist()
countries_origin = time_series['Origin'].unique().tolist()
```

```python
time_series = time_series[time_series['Value'] !='*']
time_series['Value'] =time_series['Value'].astype(int)
time_series['Country / territory of asylum/residence'] = time_series['Country / territory of asylum/residence'].astype('str')
time_series["Origin"] = time_series['Origin'].astype('str')
time_series.drop('Population type' ,axis =1, inplace = True)
time_series
```

```python
years = time_series['Year'].unique().tolist()
origin = time_series['Origin'].unique().tolist()
by_origin = pd.DataFrame(columns=['Year', 'Country','Origin_refugees'])
for y in years:
    temp = time_series[time_series['Year']==y]
    for o in origin:
        country = temp[temp['Origin'] == o]
        by_origin = by_origin.append({'Year':y,'Country':o,'Origin_refugees':country['Value'].sum()},ignore_index=True)
by_origin
```

```python
countries = time_series['Country / territory of asylum/residence'].unique().tolist()
in_country = pd.DataFrame(columns=['Year', 'Country','in_country_refugees'])
for y in years:
    temp = time_series[time_series['Year']==y]
    for c in countries:
        country = temp[temp['Country / territory of asylum/residence'] == c]
        in_country = in_country.append({'Year':y,'Country':c,'in_country_refugees':country['Value'].sum()},ignore_index=True)
in_country
```

```python
refugees_time = pd.merge(in_country, by_origin,  how='inner', left_on=['Year','Country'], right_on = ['Year','Country'])
```
