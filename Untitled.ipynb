```python
import pandas as pd
```

```python
df  = pd.read_csv('asylum_seekers.csv')
df.drop('Unnamed: 0', axis =1 , inplace= True)
```

```python
df.rename(columns={'Country / territory of asylum/residence': 'country_of_asylum','RSD procedure type / level': 'rsd_procedure_type',
               'Tota pending start-year':'total_pending_start_year', 'of which UNHCR-assisted(start-year)':'num_unhcr_assisted_start_year' ,
                'Applied during year':'applied_during_year', 'Otherwise closed': 'otherwise_closed',
                  'Total decisions': 'total_decisions','Total pending end-year' : 'total_pending_end_year',
                   'Rejected' : 'rejected', 'Origin': 'origin','Year':'year',
                  'of which UNHCR-assisted(end-year)': 'num_unhcr_assisted_end_year'}, inplace=True)
```

```python
#filling nulls
df['rsd_procedure_type'].fillna('-', inplace = True)
```

```python
df
```

```python
for col in ['total_pending_start_year', \
            'num_unhcr_assisted_start_year', \
            'applied_during_year', \
            'decisions_recognized', \
            'decisions_other', \
            'rejected', \
            'otherwise_closed', \
            'total_decisions', \
            'total_pending_end_year', \
            'num_unhcr_assisted_end_year']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].astype('float')
df
```

```python
df.fillna(0,inplace=True)
```

```python
## making a sum dataframe of the data:
temp = df
temp.drop('rsd_procedure_type',axis =1 , inplace = True )
temp
```

```python
origins = temp['origin'].unique().tolist()
```

```python
by_origin = pd.DataFrame(columns=['origin','total_applied', 'total_recognized'])
for o in origins:
    country = temp[temp['origin'] == o]
    by_origin = by_origin.append({'origin':o,'total_applied':country['applied_during_year'].sum(), 'total_recognized':country['decisions_recognized'].sum()},ignore_index=True)
by_origin
```

```python
assylum_contries = pd.DataFrame(columns=['country_of_asylum','total_applied', 'total_recognized'])
for o in origins:
    country = temp[temp['country_of_asylum'] == o]
    assylum_contries = assylum_contries.append({'country_of_asylum':o,'total_applied':country['applied_during_year'].sum(), 'total_recognized':country['decisions_recognized'].sum()},ignore_index=True)
assylum_contries
```

```python
import pandas as pd
time_series = pd.read_csv('time_series.csv')
time_series = time_series[time_series['Year']>=1990]
time_series
```

```python
time_series['Population type'].unique().tolist()
```

```python
countries_asylum = time_series['Country / territory of asylum/residence'].unique().tolist()
countries_origin = time_series['Origin'].unique().tolist()
```

```python
time_series = time_series[time_series['Value'] !='*']
time_series['Value'] =time_series['Value'].astype(int)
time_series['Country / territory of asylum/residence'] = time_series['Country / territory of asylum/residence'].astype('str')
time_series["Origin"] = time_series['Origin'].astype('str')
time_series['Population type'] = time_series['Population type'].astype('str')
time_series.dtypes
```

```python
time_series['Population type'].unique().tolist()
```

```python
## Asylum countries, yearly
assylum_contries = pd.DataFrame(columns=['country','ِrefugees_to'])
for o in countries_asylum:
    country = time_series[time_series['Country / territory of asylum/residence'] == o]
    assylum_contries = assylum_contries.append({'country':o,'ِrefugees_to':country['Value'].sum()},ignore_index=True)
assylum_contries
```

```python
origin_contries = pd.DataFrame(columns=['country','ِrefugees_from'])
for o in countries_origin:
    country = time_series[time_series['Origin'] == o]
    origin_contries = origin_contries.append({'country':o,'ِrefugees_from':country['Value'].sum()},ignore_index=True)
origin_contries
```

```python
df_countries = pd.merge(origin_contries, assylum_contries, on='country')

df_countries
```

```python
# for every country calculate the following displace type
cols =['country', 'Internally displaced persons','Refugees (incl. refugee-like situations)', 'Returnees', 'Returned IDPs', 'Others of concern','Asylum-seekers','Stateless']
population_types = pd.DataFrame(columns=cols)
for i in countries_origin:
    for j in cols[1:]:
        country = time_series[time_series['Origin'] == i]
        internally = country.loc[time_series['Population type'] == 'Internally displaced persons', 'Value'].sum()
        refugees = country.loc[time_series['Population type'] == 'Refugees (incl. refugee-like situations)', 'Value'].sum()
        returnees = country.loc[time_series['Population type'] == 'Returnees', 'Value'].sum()
        idps = country.loc[time_series['Population type'] == 'Returned IDPs', 'Value'].sum()
        ooc = country.loc[time_series['Population type'] == 'Others of concern', 'Value'].sum()
        as_seekers = country.loc[time_series['Population type'] == 'Asylum-seekers', 'Value'].sum()
        stateless =  country.loc[time_series['Population type'] == 'Stateless', 'Value'].sum()
        population_types =population_types.append({'country':i,'Internally displaced persons':internally,
                                                  'Refugees (incl. refugee-like situations)': refugees,
                                                  'Returnees':returnees,
                                                  'Returned IDPs': idps,
                                                  'Others of concern': ooc,
                                                  'Asylum-seekers': as_seekers,
                                                  'Stateless': stateless},ignore_index=True)


population_types
```

```python
population_types.drop_duplicates(inplace=True)
population_types
```

```python
df_countries = pd.merge(df_countries, population_types, on='country')
```

```python
df_countries
df_countries.to_csv('countries.csv', index=False)
```
